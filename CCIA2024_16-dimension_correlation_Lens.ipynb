```python
barem_df = pd.read_csv("drive/MyDrive/barem.csv")
barem_df
```

```python
df_dimensions_s = df.loc[:, "EW":"RI"]

# Función para obtener la puntuación correspondiente
def get_standard_value(column, value):
    for i, row in barem_df.iterrows():
      if pd.isna(row[column]):
        continue
      if '-' in row[column]:  # Si es un rango
        low, high = map(int, row[column].split('-'))
        if low <= value <= high:
            return row['Score']
      elif row[column] != '':  # Si está vacío
        if value == int(row[column]):
          return row['Score']
    return np.nan

# Aplicar la función a cada valor en df
for column in df_dimensions_s.columns:
    print(column)
    df_dimensions_s[column] = df_dimensions_s[column].apply(lambda x: get_standard_value(column, x))

df_dimensions_s
```

```python
df_dimensions_s['QOL_s']= df_dimensions_s.sum(axis=1)
```

```python
df_dimensions_s["EW_n"] = df_dimensions_s['EW'] / 20
df_dimensions_s["IR_n"] = df_dimensions_s['IR'] /20
df_dimensions_s["MW_n"] = df_dimensions_s['MW'] /20
df_dimensions_s["PD_n"] = df_dimensions_s['PD'] /20
df_dimensions_s["PW_n"] = df_dimensions_s['PW'] /20
df_dimensions_s["SD_n"] = df_dimensions_s['SD'] /20
df_dimensions_s["SI_n"] = df_dimensions_s['SI'] / 20
df_dimensions_s["RI_n"] = df_dimensions_s['RI']/20
df_dimensions_s["QOL_sn"] = df_dimensions_s['QOL_s'] /160
df_dimensions_s_normalized =  df_dimensions_s.loc[:, "EW_n":"QOL_sn"]
```

```python
for column in df_dimensions_s.loc[:, "EW":"RI"].columns:
  df_dimensions_s[column + '_mod'] = np.select(
    [df_dimensions_s[column] < 7,
     df_dimensions_s[column].between(7, 10),
     df_dimensions_s[column].between(11, 13),
     df_dimensions_s[column].between(13, 20)],
    [1, 0, 0, 1],
    np.nan)
  df_dimensions_s[column + '_level'] = np.select(
    [df_dimensions_s[column] < 7,
     df_dimensions_s[column].between(7, 10),
     df_dimensions_s[column].between(11, 13),
     df_dimensions_s[column].between(13, 20)],
    [0, 0, 1, 1],
    np.nan)

df_dimensions_s['QOL_label'] = np.select(
    [df_dimensions_s["QOL_s"] > 115,
     df_dimensions_s["QOL_s"] < 86,
     df_dimensions_s["QOL_s"].between(86, 95),
     df_dimensions_s["QOL_s"].between(96, 105),
     df_dimensions_s["QOL_s"].between(106, 115)],
    ['very-high', 'very-low', 'low', 'medium', 'high'],
    np.nan
)

df_dimensions_s['QOL_labeln'] = np.select(
    [df_dimensions_s["QOL_s"] > 115,
     df_dimensions_s["QOL_s"] < 86,
     df_dimensions_s["QOL_s"].between(86, 95),
     df_dimensions_s["QOL_s"].between(96, 105),
     df_dimensions_s["QOL_s"].between(106, 115)],
    #[4, 0, 1, 2, 3],
    [0.9, 0.1, 0.3, 0.5, 0.7],
    np.nan
)
df_dimensions_s
```

```python
df_LENs = df_dimensions_s.loc[:, "EW":"RI"]
df_LENs['QOL'] = df_dimensions_s['QOL_s']

def get_standard_value(column, value):
    for i, row in barem_df.iterrows():
      if pd.isna(row[column]):
        continue
      if '-' in row[column]:
        low, high = map(int, row[column].split('-'))
        if low <= value <= high:
            return row['Score']
      elif row[column] != '':
        if value == int(row[column]):
          return row['Score']
    return np.nan

for column in df_LENs.columns:
    print(column)
    df_LENs[column + '_vlow'] = np.zeros(len(q_df))
    df_LENs[column + '_low'] = np.zeros(len(q_df))
    df_LENs[column + '_med'] = np.zeros(len(q_df))
    df_LENs[column + '_high'] = np.zeros(len(q_df))
    df_LENs[column + '_vhigh'] = np.zeros(len(q_df))
```

```python
import torch
import torch_explain as te
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from torch_explain.nn.functional import prune_equal_fanin
from torch_explain.logic.nn import entropy
from sklearn.metrics import accuracy_score
from torch.nn.functional import one_hot
```

```python
def init_model():
  return torch.nn.Sequential(
    te.nn.EntropyLinear(c_traint.shape[1], 120, n_classes=1),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(120, 69),
    torch.nn.LeakyReLU(),
    torch.nn.Linear(69, 1),
)
```

```python
def train_model(model, c, y):
  optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)
  loss_form = torch.nn.BCEWithLogitsLoss()
  model.train()

  for epoch in range(3500):
      optimizer.zero_grad()
      y_pred = model(c).squeeze(-1)
      loss = loss_form(y_pred.reshape(-1), y) + 0.0001 * te.nn.functional.entropy_logic_loss(model)
      loss.backward()
      optimizer.step()
  return model
```

```python
def make_explanation_readable(explanation, concept_list):
  for i, concept in enumerate(concept_list):
    if i < 10:
      replace_string = 'feature000000000' + str(i)
    else:
      replace_string = 'feature00000000' + str(i)

    explanation = explanation.replace(replace_string, concept)
  return explanation
```
